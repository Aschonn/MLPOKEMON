```python
#This cell looks at all of the .jpg files in the directory this notebook is in.
#It then makes a copy of each .jpg file as a .png file.

# I acknowledge this script could be improved greatly, but for our purposes I think it'll do.





import fnmatch
import os
from PIL import Image

directory_path = os.getcwd()

fileList = [f for f in os.listdir(str(directory_path)) if fnmatch.fnmatch(f, '*.jpg')]

for file in fileList:
    fileSeparator = file.split(".jpg")
    img = Image.open(str(fileSeparator[0]) + ".jpg")
    img.save(str(fileSeparator[0]) + ".png")
    print(str(file) + " Has been saved as .png")


```

```python
# This cell looks at all of the .png files in the directory and renames them consecutively based
# on the "pokeVar" variable. 

import fnmatch
import os
directory_path = os.getcwd()

fileList = [f for f in os.listdir(str(directory_path)) if fnmatch.fnmatch(f, '*.png')]


fileCount = 0




#This is the variable that will determine the file name after the script is run.
pokeVar = "Combusken"




for i in fileList:
    try:
        os.rename(str(i), str(pokeVar) + "_" + str(fileCount) + ".png")
        fileCount += 1
        print(str(i) + " Has been renamed.")
    except:
        print("Unable to rename file. Destination file likely already exists. Moving to next file.")
```
